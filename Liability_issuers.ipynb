### Pobranie danych z XML

```python
tree = ET.parse(r'....\redadbt00101_2022-09-23.xml')
```

```python
def info_kdpw(tree):
    global kdpw_info
    kdpw_info = []


    root = tree.getroot()
    for iss in root.findall('{urn:kdpw:xsd:reda.dbt.001.01}DbtRpt'):
        for cos in iss.findall('{urn:kdpw:xsd:reda.dbt.001.01}IssrDbtRpt'):
            for lei, quant in zip(cos.iter('{urn:kdpw:xsd:reda.dbt.001.01}IssrLEI'), cos.iter('{urn:kdpw:xsd:reda.dbt.001.01}TotDbtAmt')):
                lei = lei.text        
                quantity = quant.text
                kdpw_info.append([lei,quantity])

    return kdpw_info
```

```python
kdpw_info = info_kdpw(tree)
```

### Pobranie podmiotów z kodem LEI

```python

list_issuers = []
cursor.execute("""SELECT id,LEI FROM tam_podmiot WHERE LEI is not null""")
for row in cursor:
    issuer = [int(row[0]), row[1]]
    list_issuers.append(issuer)

list_issuers
```

### Pobieranie info o długu

```python
def debt_kdpw(lei):
    global quantity
    for i in kdpw_info:
        if i[0] == lei:
            quantity = i[1]
        else:
            continue

        return quantity
```
